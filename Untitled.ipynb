```python
def convertToTensor(filename_tensor):
    image = tf.image.decode_jpeg(filename_tensor)
    image = tf.cast(image, tf.float32)
    width = image.shape[0]
    height = image.shape[1]
    div = tf.fill(image.shape, (255.0))
    image = tf.math.truediv(image, div)
    return image, width, height


def trim(imageTensor, maxHeigth, maxWidth, meanHeight, meanWidth):
    imageTensor = tf.image.resize_with_pad(imageTensor, meanHeight, meanWidth, method=tf.image.ResizeMethod.BILINEAR)
    print(imageTensor.shape)
    return

def roundup(x):
    return int(math.ceil(x / 10.0)) * 10
    
    
#     paddings = tf.constant([[1, 1,], [2, 2]])
#     tf.pad(imageTensor, [int(totalHeight/count),int(totalWidth/count), 3])
    
```
